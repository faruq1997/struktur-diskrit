## **I. KUANTOR (Quantifier)**

### **1. Definisi Kuantor Pernyataan**

Kuantor adalah simbol logika yang digunakan untuk menyatakan **jumlah atau cakupan** dari objek yang terlibat dalam suatu pernyataan.
Dengan kata lain, kuantor menunjukkan **seberapa banyak** elemen dalam semesta pembicaraan (domain) yang memenuhi suatu sifat tertentu.

Dalam logika predikat, pernyataan biasanya berbentuk:

> "Untuk semua x, berlaku sifat P(x)"
> atau
> "Ada x yang memiliki sifat P(x)"

Misalkan P(x) adalah pernyataan (predikat) yang melibatkan variabel x, dan D adalah himpunan semesta pembicaraan (domain).

Maka P disebut fungsi **proposisi**, artinya nilai kebenarannya bergantung pada nilai x di dalam D.

Contoh:

P(x): "x adalah bilangan bulat genap"

D = {1, 2, 3, 4, 5}

Maka:

P(1): “1 adalah bilangan bulat genap” → Salah (F)

P(2): “2 adalah bilangan bulat genap” → Benar (T)

P(3): “3 adalah bilangan bulat genap” → Salah (F)

dan seterusnya.

Untuk menyatakan berapa banyak elemen D yang membuat pernyataan P(x) benar, digunakanlah kuantor (quantifier).
---

### **2. Macam-Macam Kuantor**

| Jenis Kuantor            | Simbol | Dibaca                   | Makna                                                           |
| ------------------------ | ------ | ------------------------ | --------------------------------------------------------------- |
| **Kuantor Universal**    | ∀      | “Untuk semua” / “Setiap” | Menyatakan bahwa P(x) benar untuk **seluruh** x dalam D         |
| **Kuantor Eksistensial** | ∃      | “Ada” / “Terdapat”       | Menyatakan bahwa P(x) benar untuk **setidaknya satu** x dalam D |

---

### **3. Contoh Kuantor Pernyataan**

#### a. **Kuantor Universal (∀)**

Pernyataan:

> "Semua bilangan genap habis dibagi 2"

Ditulis dalam logika predikat:

> ∀x ∈ ℤ, jika x genap maka x habis dibagi 2.
> ∀x (Genap(x) → DapatDibagi(x, 2))

**Makna:**
Tidak ada satu pun bilangan genap yang tidak habis dibagi 2.

---

#### b. **Kuantor Eksistensial (∃)**

Pernyataan:

> "Ada bilangan ganjil yang lebih besar dari 10."

Ditulis dalam logika predikat:

> ∃x ∈ ℤ, (Ganjil(x) ∧ x > 10)

**Makna:**
Minimal ada satu bilangan ganjil (misalnya 11, 13, dst.) yang memenuhi kondisi tersebut.

---

Contoh lain, bayangkan kamu punya **sebuah kotak berisi pulpen atau spidol** dengan warna berbeda (merah, biru, hitam, hijau, dst.), dan kamu ingin membuat pernyataan tentang “warna yang masih bisa menulis”.

#### 🖊️ a. Kuantor Universal (∀)

> “Semua spidol di kotak ini masih bisa menulis.”

Artinya:
Untuk **setiap spidol** yang kamu ambil dari kotak, ketika diuji, **semuanya menulis dengan baik**.

Jika ternyata ada **satu saja spidol** yang tintanya kering, maka pernyataan **“semua spidol masih bisa menulis”** otomatis **salah**.

Dalam logika:

> ∀x ∈ D, P(x)
> Artinya: setiap x di D memenuhi sifat P(x).

---

#### 🖊️ b. Kuantor Eksistensial (∃)

> “Ada spidol di kotak ini yang masih bisa menulis.”

Artinya:
Kamu cukup menemukan **satu spidol saja** yang bisa menulis, maka pernyataan ini sudah **benar** — meskipun spidol lain mungkin kering semua.

Dalam logika:

> ∃x ∈ D, P(x)
> Artinya: ada minimal satu x dalam D yang memenuhi sifat P(x).

---

### **4. Negasi Kuantor**

Ketika kita menegasikan suatu pernyataan berkuantor, kuantornya juga berubah:

| Asli    | Negasi   | Arti Negasi                        |
| ------- | -------- | ---------------------------------- |
| ∀x P(x) | ∃x ¬P(x) | Tidak semua benar → ada yang salah |
| ∃x P(x) | ∀x ¬P(x) | Tidak ada yang memenuhi            |

**Contoh:**

* Pernyataan: ∀x, x² ≥ 0 → semua bilangan kuadrat tidak negatif.
  Negasi: ∃x, x² < 0 → ada bilangan yang kuadratnya negatif (ini salah, karena tidak ada).

---

Ini analoginya:

| Pernyataan Asli | Negasi   | Arti                                   |
| --------------- | -------- | -------------------------------------- |
| ∀x P(x)         | ∃x ¬P(x) | “Tidak semua benar” → “Ada yang salah” |
| ∃x P(x)         | ∀x ¬P(x) | “Tidak ada yang benar” → “Semua salah” |

**Contoh:**

* Pernyataan: ∀ spidol bisa menulis.
  Negasi: ∃ spidol yang tidak bisa menulis.
* Pernyataan: ∃ spidol yang bisa menulis.
  Negasi: ∀ spidol tidak bisa menulis.

---

### **5. Cara Penerapan Kuantor**

**Langkah-langkah:**

1. Tentukan **domain** atau semesta pembicaraan.
2. Tentukan **predikat** yang ingin diuji.
3. Gunakan **kuantor** yang sesuai (∀ atau ∃).
4. Lakukan **evaluasi logika** berdasarkan sifat atau definisi yang berlaku.

**Contoh Soal:**

> Tentukan kebenaran dari pernyataan:
> ∀x ∈ ℝ, x² ≥ 0

**Penyelesaian:**
Semua bilangan real yang dikuadratkan pasti hasilnya ≥ 0 → **Benar.**

**Contoh Soal Lain:**

**Soal 1:**
D = {1, 2, 3, 4, 5}, P(x): “x < 6”

* ∀x ∈ D, P(x): “Semua x < 6” → **Benar**
* ∃x ∈ D, P(x): “Ada x < 6” → **Benar**

**Soal 2:**
D = {spidol merah, biru, hijau},
P(x): “x masih bisa menulis”.

Jika hanya spidol merah dan biru yang masih menulis:

* ∀x P(x): salah (karena hijau kering)
* ∃x P(x): benar (karena merah dan biru masih bisa)

### **7. Penerapan Kuantor dalam Dunia Nyata**

* Dalam **pemrograman logika (AI dan Prolog)** untuk menyatakan aturan umum dan kondisi eksistensi.
* Dalam **basis data**, kuantor digunakan dalam pernyataan SQL seperti `ALL` dan `EXISTS`.
* Dalam **verifikasi sistem**, memastikan “semua input valid menghasilkan output aman” (∀) atau “ada kasus gagal” (∃).

---

Jadi:

| Aspek           | Kuantor Universal (∀)            | Kuantor Eksistensial (∃)           |
| --------------- | -------------------------------- | ---------------------------------- |
| Arti            | Berlaku untuk semua anggota      | Berlaku untuk minimal satu anggota |
| Contoh (spidol) | Semua spidol bisa menulis        | Ada spidol yang bisa menulis       |
| Sensitivitas    | Salah jika satu saja tidak benar | Benar jika ada satu yang benar     |
| Negasi          | ∃x ¬P(x)                         | ∀x ¬P(x)                           |

## (a) Predikat $P(x,y):\; x^2 < y+1$ (domain: $\mathbb{R}$)

1. $\displaystyle \forall x,\forall y,\; x^2 < y+1$.  
   **Salah.**  
   *Alasan:* Untuk menjadi benar harus berlaku untuk semua pasangan $(x,y)$. Pilih misalnya $x=100$ dan $y=-1000$:  
   $x^2 = 10000$ dan $y+1 = -999$, maka $10000 < -999$ salah.  
   Jadi ada pasangan yang membuatnya false → pernyataan universal palsu.

2. $\displaystyle \forall x,\exists y,\; x^2 < y+1$.  
   **Benar.**  
   *Alasan:* Untuk setiap $x$ kita harus menemukan setidaknya satu $y$ sehingga $x^2 < y+1$.  
   Misal, untuk setiap $x$, pilih $y = x^2$. Maka $y+1 = x^2 + 1$, dan jelas $x^2 < x^2 + 1$.  
   Jadi untuk setiap $x$ ada $y$ yang memenuhi.

3. $\displaystyle \exists x,\forall y,\; x^2 < y+1$.  
   **Salah.**  
   *Alasan:* Pernyataan ini menuntut ada suatu $x$ yang membuat $x^2 < y+1$ **untuk semua** $y$.  
   Tetapi untuk $x$ tetap, jika kita ambil $y \to -\infty$, maka $y+1$ bisa lebih kecil dari $x^2$.  
   Contoh: ambil $x = 0$, pilih $y = -100$ → $0 < -99$ salah.  
   Jadi tidak ada $x$ yang bekerja untuk semua $y$.

4. $\displaystyle \exists x,\exists y,\; x^2 < y+1$.  
   **Benar.**  
   *Alasan:* Hanya perlu satu pasangan. Contoh sederhana: $x=0, y=0$ → $0 < 0+1$ benar.

**Ringkasan (a):** F, T, F, T

---

## (b) Predikat $Q(x,y):\; (x<y) \to (x^2 < y^2)$ (domain: $\mathbb{R}$)

Ingat: implikasi $A \to B$ bernilai **FALSE** hanya bila $A$ true dan $B$ false. Jika $A$ false, implikasi selalu true.

1. $\displaystyle \forall x,\forall y,\; (x<y) \to (x^2 < y^2)$.  
   **Salah.**  
   *Counterexample:* ambil $x = -2,\; y = -1$.  
   Kita punya $x < y$ karena $-2 < -1$ (true), tetapi $x^2 = 4$ dan $y^2 = 1$, sehingga $4 < 1$ (false).  
   Jadi implikasi: true → false → keseluruhan false.  
   Karena ada pasangan yang membuatnya false, pernyataan universal palsu.

2. $\displaystyle \forall x,\exists y,\; (x<y) \to (x^2 < y^2)$.  
   **Benar.**  
   *Alasan:* Untuk setiap $x$ kita cukup menemukan satu $y$ sehingga implikasi bernilai true.  
   Cara mudah: pilih $y = x$. Maka antecedent $x < y$ menjadi false (karena $x < x$ salah),  
   sehingga implikasi otomatis true.  
   Jadi untuk setiap $x$ ada (misalnya $y = x$) yang membuat implikasi benar.

3. $\displaystyle \exists x,\forall y,\; (x<y) \to (x^2 < y^2)$.  
   **Benar.**  
   *Alasan & contoh:* Cukup tunjukkan satu nilai $x$ yang bekerja untuk semua $y$.  
   Ambil $x = 0$. Untuk semua $y$:
   - Jika $0 < y$, maka $y > 0$ sehingga $y^2 > 0$ dan $0 < y^2$ → consequent true.
   - Jika $0 < y$ false (yaitu $y \leq 0$), maka antecedent false → implikasi true.  
   Jadi untuk setiap $y$, implikasi terpenuhi. Maka $\exists x\,\forall y$ benar (contoh: $x = 0$).

   *(Catatan: nilai lain juga mungkin bekerja, misalnya setiap $x \leq 0$ — tetapi tidak perlu diverifikasi semua; cukup satu contoh.)*

4. $\displaystyle \exists x,\exists y,\; (x<y) \to (x^2 < y^2)$.  
   **Benar.**  
   *Alasan:* Banyak pasangan memenuhi. Contoh: $x = 1,\; y = 2$:  
   $1 < 2$ benar dan $1^2 < 2^2$ (yaitu $1 < 4$) juga benar → implikasi true.

**Ringkasan (b):** F, T, T, T

**(a)** $\displaystyle (\forall x\, P(x)) \land (\exists y\, Q(y))$  
*(Interpretasi wajar dari: “Semua $x$ memenuhi $P$, dan ada $y$ yang memenuhi $Q$.")*

1. **Negasi:**  
   $\displaystyle \neg\big[(\forall x\, P(x)) \land (\exists y\, Q(y))\big]$

2. **Hukum De Morgan:**  
   $\displaystyle \neg(\forall x\, P(x)) \,\lor\, \neg(\exists y\, Q(y))$

3. **Negasi kuantor:**  
   $\displaystyle (\exists x\, \neg P(x)) \,\lor\, (\forall y\, \neg Q(y))$

**Bentuk akhir:**  
$\displaystyle \exists x\, \neg P(x) \;\lor\; \forall y\, \neg Q(y)$

**Terjemahan dalam bahasa alami:**  
> “Ada suatu $x$ yang tidak memenuhi $P$, **atau** semua $y$ tidak memenuhi $Q$.”

---

**(b)** $\displaystyle (\exists x\, P(x)) \lor (\forall y\, Q(y))$  
*(Interpretasi: “Ada $x$ yang memenuhi $P$, atau semua $y$ memenuhi $Q$.")*

1. **Negasi:**  
   $\displaystyle \neg\big[(\exists x\, P(x)) \lor (\forall y\, Q(y))\big]$

2. **Hukum De Morgan:**  
   $\displaystyle \neg(\exists x\, P(x)) \,\land\, \neg(\forall y\, Q(y))$

3. **Negasi kuantor:**  
   $\displaystyle (\forall x\, \neg P(x)) \,\land\, (\exists y\, \neg Q(y))$

**Bentuk akhir:**  
$\displaystyle \forall x\, \neg P(x) \;\land\; \exists y\, \neg Q(y)$

**Terjemahan dalam bahasa alami:**  
> “Semua $x$ tidak memenuhi $P$, **dan** ada suatu $y$ yang tidak memenuhi $Q$.”

---

**(c)** $\displaystyle \forall x\, \exists y\, \big(P(x) \,\vee\, Q(y)\big)$  
*(Interpretasi: “Untuk setiap $x$, ada $y$ sehingga $P(x)$ atau $Q(y)$ benar.")*

1. **Negasi:**  
   $\displaystyle \neg\big[\forall x\, \exists y\, (P(x) \vee Q(y))\big]$

2. **Negasi kuantor luar:**  
   $\displaystyle \exists x\, \neg\big[\exists y\, (P(x) \vee Q(y))\big]$

3. **Negasi kuantor dalam:**  
   $\displaystyle \exists x\, \forall y\, \neg(P(x) \vee Q(y))$

4. **Hukum De Morgan (logika proposisional):**  
   $\displaystyle \neg(P(x) \vee Q(y)) \equiv \neg P(x) \land \neg Q(y)$

**Bentuk akhir:**  
$\displaystyle \exists x\, \forall y\, \big(\neg P(x) \,\land\, \neg Q(y)\big)$

**Terjemahan dalam bahasa alami:**  
> “Ada suatu $x$ sedemikian sehingga untuk **setiap** $y$, baik $P(x)$ salah **maupun** $Q(y)$ salah.”

```python
# Kampus ingin memeriksa data mahasiswa dari sistem akademik:
# 1. Apakah semua mahasiswa sudah memiliki email kampus (@nusamandiri.ac.id)
# 2. Apakah ada minimal satu mahasiswa berprestasi (status = 'aktif' dan ipk ≥ 3.75)

# --- Data Mahasiswa ---
mahasiswa = [
    {"nama": "Rina", "email": "11250001@nusamandiri.ac.id", "status": "aktif", "ipk": 3.8},
    {"nama": "Andi", "email": "11250021@nusamandiri.ac.id", "status": "aktif", "ipk": 3.91},
    {"nama": "Siti", "email": "siti@gmail.com", "status": "aktif", "ipk": 3.7},
]

# --- 1. Kuantor Universal (∀): Semua mahasiswa punya email kampus ---
semua_email_kampus = True  # Asumsi awal semua benar

for m in mahasiswa:
    # Cek apakah email mahasiswa mengandung domain kampus
    if "@nusamandiri.ac.id" not in m["email"]:
        semua_email_kampus = False
        break  # hentikan loop kalau ada yang tidak sesuai

# --- 2. Kuantor Eksistensial (∃): Ada mahasiswa berprestasi (ipk >= 3.75) ---
ada_mahasiswa_berprestasi = False

for m in mahasiswa:
    if m["status"] == "aktif" and m["ipk"] >= 3.75:
        ada_mahasiswa_berprestasi = True
        break  # cukup satu saja sudah memenuhi

# --- 3. Tampilkan Hasil ---
print("Apakah semua mahasiswa punya email kampus?", semua_email_kampus)
print("Apakah ada mahasiswa berprestasi?", ada_mahasiswa_berprestasi)
```


## **II. INDUKSI MATEMATIKA (Mathematical Induction)**

### **1. Definisi**

Induksi matematika adalah **metode pembuktian** untuk menunjukkan bahwa suatu pernyataan benar untuk **semua bilangan bulat positif (n ≥ 1)**.

Konsepnya mirip dengan **rangkaian domino**:

> Jika suatu pernyataan benar untuk langkah awal, dan jika kebenaran pada langkah ke-k menjamin kebenaran pada langkah ke-(k+1), maka pernyataan itu benar untuk semua langkah berikutnya.

---

### **2. Syarat Pembuktian Induksi**

Untuk membuktikan pernyataan P(n) benar untuk semua n ≥ k (biasanya k = 1), kita harus menunjukkan:

1. **Basis (Langkah Dasar):** P(k) benar.
2. **Langkah Induksi:** Jika P(m) benar, maka P(m+1) juga benar.

Jika kedua langkah itu terbukti, maka P(n) benar untuk semua n ≥ k.

---

### **3. Langkah-Langkah Pembuktian**

| Langkah       | Keterangan                                   |
| ------------- | -------------------------------------------- |
| **Langkah 1** | Buktikan bahwa P(1) benar (basis induksi)    |
| **Langkah 2** | Asumsikan P(k) benar (hipotesis induksi)     |
| **Langkah 3** | Gunakan P(k) untuk membuktikan P(k+1) benar  |
| **Langkah 4** | Simpulkan bahwa P(n) benar untuk semua n ≥ 1 |

---

### **4. Contoh Soal dan Penyelesaian**

**Soal:**
Buktikan dengan induksi matematika bahwa:
1 + 2 + 3 + ... + n = n(n + 1)/2

**Penyelesaian:**

#### Langkah 1: Basis Induksi

Untuk n = 1:
1 = 1(1 + 1)/2 = 1 → **Benar**

#### Langkah 2: Hipotesis Induksi

Misalkan P(k) benar, yaitu:
1 + 2 + 3 + ... + k = k(k + 1)/2

#### Langkah 3: Langkah Induktif

Buktikan bahwa P(k+1) benar:
1 + 2 + 3 + ... + k + (k+1)
= [k(k + 1)/2] + (k + 1)
= (k + 1)(k/2 + 1)
= (k + 1)(k + 2)/2

→ P(k+1) **terbukti benar**

#### Kesimpulan:

Karena basis dan langkah induksi terbukti benar, maka
**1 + 2 + 3 + ... + n = n(n + 1)/2** benar untuk semua n ≥ 1.

---

### **5. Contoh Lain**

Buktikan bahwa untuk semua n ≥ 1,
2ⁿ ≥ n + 1

#### Pembuktian:

* **Basis:** n=1 → 2¹=2 ≥ 2 → benar
* **Hipotesis:** misal 2ᵏ ≥ k+1
* **Langkah Induktif:**
  2ᵏ⁺¹ = 2×2ᵏ ≥ 2(k+1) = 2k+2 ≥ (k+1)+1 (karena k≥1)
  → benar
  Jadi 2ⁿ ≥ n + 1 untuk semua n ≥ 1.

---

### **6. Analogi Sederhana**

Bayangkan tangga tanpa akhir:

* Jika kamu bisa berdiri di anak tangga pertama (basis),
* Dan dari satu anak tangga kamu bisa naik ke anak tangga berikutnya (langkah induksi),
* Maka kamu bisa mencapai **semua anak tangga**.

---

### **7. Penerapan Induksi Matematika dalam Dunia Nyata**

* **Algoritma rekursif:** Pembuktian kebenaran algoritma dilakukan dengan prinsip induksi.
* **Komputer & pemrograman:** Pembuktian benar/tidaknya fungsi rekursif.
* **Rancangan rangkaian listrik:** Menunjukkan kestabilan sistem n-komponen berdasarkan sistem n−1.
* **Kriptografi:** Pembuktian pola keamanan enkripsi.

## **RANGKUMAN SINGKAT**

| Konsep                       | Inti                | Simbol/Rumus      | Aplikasi                         |
| ---------------------------- | ------------------- | ----------------- | -------------------------------- |
| **Kuantor Universal (∀)**    | Berlaku untuk semua | ∀x P(x)           | Validasi semua kondisi           |
| **Kuantor Eksistensial (∃)** | Ada minimal satu    | ∃x P(x)           | Pengecekan eksistensi            |
| **Negasi Kuantor**           | Membalik ∀ ↔ ∃      | ¬∀xP(x) ≡ ∃x¬P(x) | Kontradiksi logika               |
| **Induksi Matematika**       | Pembuktian bertahap | Basis + Induksi   | Pembuktian formula dan algoritma |

```python
# Sistem akademik Universitas Nusa Mandiri ingin menghitung total SKS yang diambil oleh seorang mahasiswa selama beberapa semester.
# Tiap semester rata-rata mahasiswa mengambil 20 SKS.

# Di sini akan dibuktikan bahwa fungsi rekursif yang menghitung total SKS ini selalu benar untuk semua n semester (n ≥ 1) menggunakan konsep induksi.

# --- Fungsi Rekursif untuk Menghitung Total SKS ---
def total_sks(semester):
    if semester == 1:
        return 20  # semester pertama ambil 20 SKS
    else:
        return 20 + total_sks(semester - 1)

# --- Uji Fungsi ---
for n in range(1, 6):  # uji 5 semester pertama
    print(f"Total SKS selama {n} semester = {total_sks(n)}")

# --- Verifikasi Induksi ---
# Rumus teori: total SKS = 20 * semester
def verifikasi_induksi(semester):
    hasil_rekursif = total_sks(semester)
    hasil_rumus = 20 * semester
    return hasil_rekursif == hasil_rumus

# Cek apakah fungsi benar untuk semua semester dari 1 sampai 10
semua_benar = all(verifikasi_induksi(n) for n in range(1, 11))
print("Apakah fungsi rekursif sesuai dengan rumus teori untuk semua n?", semua_benar)
```

```python
# --- Data Mahasiswa (Simulasi dari Sistem Akademik Nusa Mandiri) ---
mahasiswa = [
    {"nama": "Rina", "ipk": 3.85, "semester": 8},
    {"nama": "Andi", "ipk": 2.95, "semester": 6},
    {"nama": "Siti", "ipk": 3.20, "semester": 8},
]

# ---------------------------------------------------------
# Bagian 1️⃣ : Kuantor Universal (∀) → Semua IPK harus valid
# ---------------------------------------------------------
semua_ipk_valid = all(0 <= m["ipk"] <= 4 for m in mahasiswa)

print("Apakah semua mahasiswa memiliki IPK valid (0–4)?", semua_ipk_valid)

# ---------------------------------------------------------
# Bagian 2️⃣ : Kuantor Eksistensial (∃) → Ada mahasiswa layak lulus
# ---------------------------------------------------------
def total_sks(semester):
    """Hitung total SKS secara rekursif (20 SKS per semester)."""
    if semester == 1:
        return 20
    else:
        return 20 + total_sks(semester - 1)

# Cek apakah ada mahasiswa yang memenuhi syarat kelulusan
ada_lulus = any(m["ipk"] >= 3.0 and total_sks(m["semester"]) >= 144 for m in mahasiswa)
print("Apakah ada mahasiswa yang sudah layak lulus?", ada_lulus)

# ---------------------------------------------------------
# Bagian 3️⃣ : Induksi → Verifikasi fungsi total_sks untuk semua semester
# ---------------------------------------------------------
def verifikasi_induksi(n):
    # Rumus teori total SKS = 20 * n
    return total_sks(n) == 20 * n

# Verifikasi untuk 10 semester pertama
hasil_verifikasi = all(verifikasi_induksi(i) for i in range(1, 11))
print("Apakah fungsi total_sks terbukti benar secara induksi?", hasil_verifikasi)
```
